# Exploring Datasets with pandas

```python
df_can = pd.read_excel('./dataset/Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
```

```python
df_can.head()
```

```python
df_can.info()
```

```python
df_can.shape
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName': 'Region'}, inplace=True)
```

```python
df_can.head()
```

```python
df_can['total'] = df_can.sum(axis=1)
```

```python
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
df_can.loc['Indonesia']
```

```python
df_can.loc['Indonesia',[2000,2009,2010,2011,2013]]
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(1980,2014)))
years
```

```python
df_can.head()
```

## Visualizing Data using Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print(plt.style.available)
```

```python
mpl.style.use(['ggplot'])
```

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

### Line Plot

```python
haiti.plot()
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
haiti.index = haiti.index.map(int)
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.text(2000,6000, '2010 Earthquake')

plt.show()
```

```python
China_India = df_can.loc[['China', 'India'],years]
China_India
```

```python
China_India.plot(kind='line')
```

```python
df_CI = China_India.transpose()
```

```python
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title('Immigration from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_can.sort_values(by='total', ascending=False, axis=0, inplace=True)
```

```python
df_top5 = df_can.head(5)
df_top5
```

```python
df_top5 = df_top5[years].transpose()
```

```python
df_top5.plot(kind='line', figsize=(14,8))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

### Area Plots

```python
df_top5.head()
```

```python
df_top5.plot(kind='area',figsize=(20, 10), 
            stacked=False,
            alpha=0.5)

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

### Histogram

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
```

```python
df_can['2013'].plot(kind='hist', figsize=(20,10))
plt.title('History of Immigration in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrant')

plt.show()
```

```python
df_can['2013'].plot(kind='hist', figsize=(11,6),xticks=bin_edges)
plt.title('History of Immigration in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrant')

plt.show()
```

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
```

```python
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind='hist', 
         figsize=(10,6),
         alpha=0.6, 
         color=['coral', 'darkslateblue', 'mediumseagreen'],
         xticks=bin_edges,
         bins=15)

plt.title('History of Immigration Denmark, Norway, Sweden')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrant')

plt.show()
```

### Bar Chart

```python
df_iceland =df_can.loc['Iceland', years]
df_iceland.head()
```

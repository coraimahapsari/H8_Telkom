# Exploring Datasets with pandas

```python
df_can = pd.read_excel('./dataset/Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
```

```python
df_can.head()
```

```python
df_can.info()
```

```python
df_can.shape
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName': 'Region'}, inplace=True)
```

```python
df_can.head()
```

```python
df_can['total'] = df_can.sum(axis=1)
```

```python
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
df_can.loc['Indonesia']
```

```python
df_can.loc['Indonesia',[2000,2009,2010,2011,2013]]
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(1980,2014)))
years
```

```python
df_can.head()
```

## Visualizing Data using Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print(plt.style.available)
```

```python
mpl.style.use(['ggplot'])
```

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

### Line Plot

```python
haiti.plot()
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
haiti.index = haiti.index.map(int)
```

### Area Plots

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
```

```python
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind='hist', 
         figsize=(10,6),
         alpha=0.6, 
         color=['coral', 'darkslateblue', 'mediumseagreen'],
         xticks=bin_edges,
         bins=15)

plt.title('History of Immigration Denmark, Norway, Sweden')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrant')

plt.show()
```

### Pie Chart

```python
df_continents = df_can.groupby('Continent', axis =0).sum()
df_continents.head()
```

```python
df_continents['total'].plot(kind='pie',
                           autopct='%1.1f%%',
                           figsize=(10, 6),
                           startangle=90,
                           shadow=True)
plt.title('Immigration to Canada by Continent')
plt.axis('equal')

plt.show()
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['total'].plot(kind='pie',
                           autopct='%1.1f%%',
                           figsize=(15, 6),
                           startangle=90,
                           shadow=True,
                            counterclock=False,
                           label=None,
                           pctdistance = 1.12,
                           colors=color_list,
                           explode=explode_list)
plt.title('Immigration to Canada by Continent', y=1.12)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['2013'].plot(kind='pie',
                           autopct='%1.1f%%',
                           figsize=(15, 6),
                           startangle=90,
                           shadow=True,
                            counterclock=False,
                           label=None,
                           pctdistance = 1.12,
                           colors=color_list,
                           explode=explode_list)
plt.title('Immigration to Canada by Continent in 2013', y=1.12)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

### Subplots

```python
fig = plt.figure()

ax0 = fig.add_subplot(1 , 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

df_CI.plot(kind='box', figsize = (20,6), 
           color='skyblue', 
           vert=False, 
           ax=ax0)
ax0.set_title('Box Plot of China and India')
ax0.set_xlabel('Number of Immigrant')
ax0.set_ylabel('Contries')

df_CI.plot(kind='line', figsize = (20,6),
          ax=ax1)
ax1.set_title('Line Plot of China and India')
ax1.set_xlabel('Number of Immigrant')
ax1.set_ylabel('Contries')

plt.show()
```

### Scaterplots

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))
```

```python
df_tot.index = map(int, df_tot.index)
```

```python
df_tot.reset_index(inplace=True)
```

```python
df_tot.columns = ['year', 'total']
```

```python
df_tot.head()
```

```python
df_tot.plot(kind='scatter',
           x='year',
           y='total',
           figsize=(10,6),
           color='darkblue')

plt.title('Total immigration to Canada')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
